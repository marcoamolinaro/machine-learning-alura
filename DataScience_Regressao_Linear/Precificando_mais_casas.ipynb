### **Lendo o DataFrame com as casas a precificar**

```python
import pandas as pd
import numpy as np
```

```python
# Lendo a base das casas a precificar
casas = pd.read_csv('/content/casas_a_precificar.csv')
casas.head()
```

```python
# Ajustando a coluna dos dados
casas.columns=['log_area_primeiro_andar', 'existe_segundo_andar', 'log_area_quintal', 'log_dist_metro']
```

```python
# Calculando o logaritmo das variaveis
casas['log_area_primeiro_andar'] = np.log(casas['log_area_primeiro_andar'])
casas['log_area_quintal'] = np.log1p(casas['log_area_quintal'])
casas['log_dist_metro'] = np.log(casas['log_dist_metro'])
```

```python
casas
```

### **Importando o modelo salvo em pickle**

```python
import pickle
```

```python
# Importando o modelo
m = open('/content/modelo_casas_a_precificar.pkl', 'rb')
modelo = pickle.load(m)
m.close()
```

### **Estimando o preço das casas**

```python
# Prevendo os valores das casas
previsoes = modelo.predict(casas)
previsoes
```

```python
# Exibindo os valores em reais
for i, previsao in enumerate(previsoes):
    print(f'Casa {i+1}: R$ {np.exp(previsao):,.2f}')
```
